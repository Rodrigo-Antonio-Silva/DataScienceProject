1) Faça uma função que dado um vetor de inteiros retorne o maior inteiro do vetor.

```python
def maior_vetor(vetor):
  if len(vetor) == 0:
    return None
  maior = vetor[0]
  for i in range(1, len(vetor)):
    if vetor[i] > maior:
      maior = vetor[i]
  return maior
```

```python
print(maior_vetor([1, 2, 3, 4, 675, 56, 7, 8, 9, 10]))
```

2) Você está organizando uma lista de alunos por turmas em uma escola. Cada turma contém vários alunos, e cada aluno é deve conter um nome, idade e id.

Exercício da aula passada, porém deve ser resolvido utilizando o que aprendemos sobre POO.

a) Crie duas turmas, cada uma contendo dois alunos.

b) Adicionar um novo aluno: Adicione um novo aluno em uma das turmas.

c) Remover um aluno: Remova um aluno de uma das turmas.

d) Exibir os alunos de uma turma

```python
class Aluno:
  def __init__(self, nome, idade, id):
    self.nome = nome
    self.idade = idade
    self.id = id
```

```python
Aluno1 = Aluno("João", 18, 1)
Aluno2 = Aluno("Maria", 19, 2)
```

```python
class Turma():
  def __init__(self, nome):
    self.nome = nome
    self.alunos = []

  def adicionar_aluno(self, aluno):
    self.alunos.append(aluno)

  def remover_aluno(self, id):
    for aluno in self.alunos:
      if aluno.id == id:
        self.alunos.remove(aluno)

  def exibir_alunos(self):
    if len(self.alunos) == 0:
      print("Não há alunos na turma")
      return
    for aluno in self.alunos:
      print(f"Aluno: {aluno.nome}, Idade: {aluno.idade}, Id: {aluno.id}")

```

```python
TurmaA = Turma("TurmaA")
TurmaB = Turma("TurmaB")

Aluno1 = Aluno("João", 18, 1)
Aluno2 = Aluno("Maria", 19, 2)

TurmaA.adicionar_aluno(Aluno1)
TurmaA.adicionar_aluno(Aluno2)

TurmaB.adicionar_aluno(Aluno1)
TurmaB.adicionar_aluno(Aluno2)
```

```python
print("Alunos turma A:")
TurmaA.exibir_alunos()
print("\nAlunos turma B:")
TurmaB.exibir_alunos()
```

```python
TurmaA.remover_aluno(1)
print("\nAlunos turma A:")
TurmaA.exibir_alunos()
```

3) Crie uma função que dado dois vetores retorne um novo vetor que é composto dos maiores elementos de cada vetor

```python
def maiores(vetor1, vetor2):
  vetor3 = []
  vetor3.append(maior_vetor(vetor1))
  vetor3.append(maior_vetor(vetor2))
  return vetor3
```

```python
print(maiores([1, 3, 400, ], [23, 627, 78]))
```

4) Crei uma função que dado dois vetores retorne um novo vetor que é a concatenação de ambos os vetores.

```python
def concatenar(vetor1, vetor2):
  vetor3 = []
  if len(vetor1) == 0 and len(vetor2) == 0:
    return "O vetores vazios"
  else:
    vetor3 = vetor1 + vetor2
    return sorted(vetor3)
```

5) Crie uma função que receba uma lista de palavras e retorne o número de palavras que tenham um tamanho maior do que 5

```python
def maior_que_5(lista):
  contador = 0
  for palavra in lista:
    if len(palavra) > 5:
      contador += 1
  return contador
```

```python
maior_que_5(['Tropeço', 'olheiro', 'cafézinho', 'juju'])
```

6) Agora crie uma função que dado um texto e um número retorne o número de palavras que tenham um tamanho maior do que o número fornecido.

```python
def palavras(t, n):
  lista = t.split()
  contador = 0
  for palavra in lista:
    if len(palavra) > n:
      contador += 1
  return contador



t = "A baba é sinistra."
n = int(input("Digite um número: "))
print(palavras(t, n))
```
