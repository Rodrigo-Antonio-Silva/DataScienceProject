```python
import pandas as pd
```

```python
import numpy as np
```

```python
df1 = pd.read_excel('/content/drive/MyDrive/Bases_DS/Análise de Dados Vale_Online_10_2020.xlsx', sheet_name=3)
```

```python
df2 = pd.read_excel('/content/drive/MyDrive/Bases_DS/Análise de Dados Vale_Online_10_2020.xlsx', sheet_name=4)
```

```python
df1.head(10)
```

```python
df2.head()
```

```python
latas = pd.merge(df2, df1, how='inner')
```

```python
latas.head()
```

```python
latas.tail()
```

```python
latas.info()
```

```python
latas.Embalagem.unique()
```

```python
latas['Embalagem'] = latas.Embalagem.astype('category')
```

```python
latas.info()
```

```python
latas.Mercado.unique()
```

```python
latas['Mercado'] = latas.Mercado.astype('category')
```

```python
latas.info()
```

```python
latas.Cliente.unique()
```

```python
latas['Cliente'] = latas.Cliente.astype('category')
```

```python
latas.info()
```

```python
latas.head()
```

```python
latas.drop('Cod Cliente', axis=1, inplace=True)
```

```python
latas.head()
```

```python
latas.rename(columns={'Unidades (000)':'Unid'}, inplace=True)
```

```python
latas.head()
```

```python
latas['Unid'] = latas.Unid.astype('int')
```

```python
latas.head()
```

```python
latas['Ano'] = pd.to_datetime(latas['Data'], format='%y-%m-%d').dt.year
```

```python
latas.head()
```

```python
pd.crosstab(latas.Ano, latas.Cliente, margins=True, normalize=True)
```

```python
pd.crosstab(latas.Cliente, latas.Ano, values=latas.Unid, aggfunc='mean', normalize='columns')
```

```python
pd.crosstab(latas.Mercado, latas.Embalagem, values=latas.Unid, aggfunc='sum', normalize='columns')
```

```python
pd.crosstab(latas.Ano, [latas.Embalagem, latas.Data.dt.quarter], values=latas.Unid, aggfunc='sum', colnames=['Embalagem', 'Quarter'], margins=True, margins_name='Total')
```

```python
pd.crosstab(latas.Ano, latas.Data.dt.quarter, values=latas.Unid, aggfunc='sum', colnames=['Trim'], normalize='index')
```
