Diariamente, para um período de 100 dias consecutivos, foi observado o número de clientes que entram numa padaria/confeitaria, a cada intervalo de uma hora, das 7:00 às 19:00 horas.


Sabendo que o tempo médio de atendimento de cada cliente é de 3 minutos e que, para garantir um nível de serviço adequado, a ocupação média de um atendente não deva ser superior a 75 % do seu tempo disponível, dimensionar o número mínimo de atendentes da padaria por faixa horária.


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/drive/MyDrive/Bases_DS/Análise de Dados Vale_Online_10_2020.xlsx',sheet_name=16)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(labels='Dia', axis=1, inplace=True)
```

```python
df.head()
```

```python
df_media_aten = np.mean(df, axis=0)
```

```python
df_media_aten
```

O atendente não deve ser superior a 45 min


```python
max_clientes = 45 / 3
```

```python
df_num = df_media_aten // max_clientes
```

```python
df_num
```

```python
%matplotlib inline
```

```python
df_num.plot.line(x='horas', y='atendente')
```
