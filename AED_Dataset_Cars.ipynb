```python
# importando os pacotes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = "white")
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# criando dataset
df = pd.read_csv(r"C:\Users\FamiliaSilva\Documents\datasets\cars_dataset.csv")
```

```python
# Visualizando as 10 primeiras linhas do dataset
df.head(10)
```

```python
# Copiando o df original
df1 = df.copy()
```

```python
# Conhecendo o dataset
df1.info()
```

```python
# Detalhes estatísticos
df1.describe()
```

```python
# Eliminando linhas com motos 0.0
df_remove = df1.loc[df1['engineSize'] == 0]
```

```python
df1 = df1.drop(df_remove.index)
```

```python
# Dados estatísticos após a eliminação
df1.describe()
```

```python
# Visualizando dados faltantes
df1.isnull().sum()
```

```python
# removendo possiveis duplicidades
df1.drop_duplicates(ignore_index=True, inplace=True)
```

```python
# Dados estatísticos após a eliminação
df1.describe()
```

```python
# Informações do dataset após ajustes
df1.info()
```

```python
# Visão das últimas cinco linhas do dataset
df1.tail()
```

```python
# Quantidade de modelos por fabricante
df1['Make'].value_counts()
```

```python
make_order = df1['Make'].value_counts().index
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Make', data= df, order= make_order, palette="Greens_r")
diverging_colors = sns.color_palette("RdBu", 10)
abs_values = df['Make'].value_counts().values
ax.bar_label(container=ax.containers[0], labels=abs_values)
plt.xlabel('Fabricantes', fontsize=12)
plt.ylabel('Qtd de modelos', fontsize=12)
plt.title('Qtd de modelos por fabricante', fontsize=20)
plt.show()
```

```python
# Relação Tipo de combustível vs. Tipo de transmissão 
# insight(Veículos a disel são predominantemente Automáticos e Semi-Automáticos já os veículos manuais são movidos predominantemente a petróleo)
pd.crosstab(df1['fuelType'], df1['transmission'], margins=True, normalize='columns')
```

```python
# Veículo com menor(mpg) rendimento distância por galão
df1.loc[df1['mpg'] == df1['mpg'].min()]
```

```python
# Veículo com maior (mpg) rendimento por galão
df1.loc[df1['mpg'] == df1['mpg'].max()]
```

```python
df_model_count = df1.groupby('model').model.count().sort_values(ascending=False).head(30)
```

```python
# 30 Principais modelos de veículos no dataset
plt.figure(figsize=(20,10))
sns.barplot(x=df_model_count.values, y=df_model_count.index, palette= 'Reds_r')
plt.xlabel('Qtd', fontsize=12)
plt.ylabel('Modelos', fontsize=12)
plt.title('30 Principais modelos de veículos no dataset', fontsize=20)
plt.show()
```

```python
# 10 Modelos com maior numero de ocorrência no dataset e seu respectivo fabricante
df1.groupby(['model', 'Make', 'transmission', 'fuelType']).model.count().sort_values(ascending=False).head(10)
```

```python
percent_model = df1.groupby(['model', 'Make']).model.count().sort_values(ascending=False).head(10)
```

```python
# 10 Modelos com maior frequência no dataset em relação ao total de veículos
print(percent_model / df1['model'].count() * 100)
```

```python
# Análise acumulativa dos modelos como maior número de ocorrencias. 
# Insigth (Os 10 modelos com maior ocorrência representam 44 % do total de veículos no dataset)
(percent_model / df1['model'].count() * 100).cumsum(axis=0)
```

```python
import matplotlib.patches as mpatches
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x ='Make', y= 'price', data= df1, order=['audi', 'BMW', 'vw', 'skoda', 'Hyundai', 'toyota', 'Ford'])
plt.axhline(df.price.mean(), 0, 1, color='red', **{'ls': '-', 'lw': 2.0})
plt.xlabel('Fabricantes', fontsize=12)
plt.ylabel('média de preços', fontsize=12)
plt.title('Preço médio por fabricante', fontsize=20)
red_patch = mpatches.Patch(color='red', label='The red data',ls='--', lw='0.1')
plt.legend(labels=['média geral'], handles=[red_patch])
plt.show()
```

```python
# Relação Fabricante x tipo de câmbio pela média de preço
plt.figure(figsize=(10,6))
crosstb = pd.crosstab(df1.transmission, df1.Make, values=df1['price'], aggfunc='mean')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(np.floor(crosstb), annot=True, fmt=".0f", linewidths=3, cmap="RdBu", annot_kws={'fontsize':12})
plt.xlabel('Fabricantes', fontsize=12)
plt.ylabel('Tipo de transmissão', fontsize=12)
plt.title('Preço médio de um veículo por fabricante, bem como tipo de câmbio', fontsize=20)
```

```python
# Preço médio dos fabricantes com base no tipo de transmissão
plt.figure(figsize=(10,6))
sns.barplot(data=df1, x='Make', y='price', hue='transmission')
plt.xlabel('Fabricantes', fontsize=12)
plt.ylabel('Preço médio', fontsize=12)
plt.title('Preço médio dos fabricantes com base no tipo de transmissão', fontsize=20)
plt.grid(axis='y')
plt.show()
```

```python
# Preço médio dos fabricantes com base no tipo de combustível
plt.figure(figsize=(10,6))
sns.barplot(data=df1, x='Make', y='price', hue='fuelType')
plt.xlabel('Fabricantes', fontsize=12)
plt.ylabel('Preço médio', fontsize=12)
plt.title('Preço médio dos fabricantes com base no tipo de combustível', fontsize=20)
plt.grid(axis='y')
plt.show()
```

```python
# Consumo médio de combustível por fabricante
plt.figure(figsize=(10,6))
sns.barplot(x='Make', y='mpg', data=df1, hue='fuelType')
plt.xlabel('Fabricantes', fontsize=12)
plt.ylabel('Consumo médio por litro', fontsize=12)
plt.title('Consumo médio de combustível por fabricante', fontsize=20)
plt.show()
```

```python
# Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
plt.figure(figsize=(10,6))
sns.barplot(x='fuelType', y='price', data=df1, hue='transmission')
plt.xlabel('Tipo de combustível', fontsize=12)
plt.ylabel('Média de preço', fontsize=12)
plt.title('Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio', fontsize=20)
plt.show()
```

```python
'''Tabela com a mediana dos valores do dataset organizadas pela mediana dos preços.
Insigth (Como a tabela comprova as observações extraídas dos gráficos acima onde 
Audi a BMW possuem os maiores preços médio por tipo de transmissão e tipo de marca)'''
df1.groupby(['Make']).median().sort_values('price', ascending=False)
```

```python
# Distribuição dos preços por fabricante com relação ao tipo de câmbio
plt.figure(figsize=(10,7))
sns.boxplot(x='Make', y='price', data=df1, hue='transmission')
plt.xlabel('Fabricantes', fontsize=12)
plt.ylabel('pPreço', fontsize=12)
plt.title('Distribuição dos preços por fabricante com relação ao tipo de câmbio', fontsize=20)
```

```python
# Distribuição do preço por  tipo de combustível com base na trasmissão
plt.figure(figsize=(10,7))
sns.boxplot(x='fuelType', y='price', data=df1, hue='transmission')
plt.xlabel('Tipo de combustível', fontsize=12)
plt.ylabel('Preço', fontsize=12)
plt.title('Distribuição do preço por  tipo de combustível com base na trasmissão', fontsize=20)
```

```python
# Distribuição fabricante por milhas por galão com base na transmissão
plt.figure(figsize=(10,7))
sns.boxplot(x='Make', y='mpg', data=df1, hue='transmission')
plt.xlabel('Fabricante', fontsize=12)
plt.ylabel('milhas por galão', fontsize=12)
plt.title('Distribuição fabricante por milhas por galão com base na transmissão', fontsize=20)
```

```python
# Distribuição Fabricante milhas por galão com base no tipo de combustível
plt.figure(figsize=(10,6))
sns.boxplot(x='Make', y='mpg', data=df1, hue='fuelType')
plt.xlabel('Fabricante', fontsize=12)
plt.ylabel('Milhas por galão', fontsize=12)
plt.title('Distribuição Fabricante milhas por galão com base no tipo de combustível', fontsize=20)
```

```python
# Porcentagem de tipos de combustível
df_fuel = df1.groupby(df1['fuelType']).model.count().sort_values(ascending=False).reset_index()
total_f = df_fuel.model.sum()
df_fuel['%'] = df_fuel.model.map(lambda x: x / total_f * 100)
df_fuel.rename(columns = {'model': 'Qtd'}, inplace=True)
```
