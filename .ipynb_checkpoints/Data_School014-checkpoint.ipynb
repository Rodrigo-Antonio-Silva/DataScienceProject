```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry', dtype={'continent':str})
```

```python
drinks.head()
```

```python
drinks.groupby('continent')['beer_servings'].mean()
```

```python
drinks[drinks['continent'] == 'Africa'].beer_servings.mean()
```

```python
drinks.groupby('continent')['beer_servings'].max()
```

```python
drinks.groupby('continent')['beer_servings'].min()
```

```python
drinks.groupby('continent')['beer_servings'].agg(['count', 'max', 'min', 'mean', 'std', 'var'])
```

```python
drinks.groupby('continent').mean()
```

```python
table = drinks.groupby('continent').mean()
```
